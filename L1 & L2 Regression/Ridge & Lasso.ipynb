RIDGE and LASSO 

Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

Dataset

```python
df=pd.read_csv("Melbourne_housing_FULL.csv")
df.head()
```

Cleaning

```python
df = df.drop(columns=["Suburb", "Address", "Method", "SellerG", "Date", "CouncilArea", "Regionname"])
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

Encoding "Type"

```python
#df = pd.get_dummies(df, columns=["Type"], drop_first=False)
df = df.astype({col: int for col in df.select_dtypes(include=["bool"]).columns})

#print(df.columns)
df.head()
```

Defining  Features(X and Y)

```python
x = df.drop(columns=["Price"])
y = df["Price"]

print(x.head(1))
```

Spliting Dataset

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Standardizing/Scaling

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

Training Ridge & Lasso

```python
ridge = Ridge(alpha=1.0)
lasso = Lasso(alpha=0.1)

ridge.fit(x_train_scaled, y_train)
lasso.fit(x_train_scaled, y_train)
```

Predictions

```python
ridge_pred = ridge.predict(x_test_scaled)
lasso_pred = lasso.predict(x_test_scaled)
```

Evaluvations using MSE

```python
ridge_mse = mean_squared_error(y_test, ridge_pred)
lasso_mse = mean_squared_error(y_test, lasso_pred)

print(ridge_mse)
print(lasso_mse)
```

Graph

```python
plt.figure(figsize=(10, 5))
plt.plot(ridge.coef_, "s-", label="Ridge")
plt.plot(lasso.coef_, "+-", label="Lasso")
plt.xlabel("Index")
plt.ylabel("Coefficient")
plt.legend()
plt.title("Ridge and Lasso Regression")
plt.show()
```
